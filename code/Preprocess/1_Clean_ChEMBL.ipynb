```python
## PROCESS 6
## Remove Outlier & Remove NAN data in [ALogP, RO5_Violations, MW]

subpc = afterprocess_5

## Remove NAN
subpc_1 = subpc[subpc.Smiles.isnull()==False]
subpc_2 = subpc_1[subpc_1.AlogP.isnull()==False]
subpc_3 = subpc_2[subpc_1.Molecular_Weight.isnull()==False]
subpc_4 = subpc_3[subpc_3.pChEMBL_Value.isnull()==False]
subpc_5 = subpc_4[~(subpc_4['AlogP']=='None')]

subpc_5[["RO5_Violations", "AlogP","Molecular_Weight"]] = subpc_5[["RO5_Violations","AlogP","Molecular_Weight"]].apply(pd.to_numeric)

## Find Statistic Border Value
Q1_RO5 = subpc_5['RO5_Violations'].quantile(0.25)
Q3_RO5 = subpc_5['RO5_Violations'].quantile(0.75)
IQR_RO5 = Q3_RO5 - Q1_RO5

Q1_LogP = subpc_5['AlogP'].quantile(0.25)
Q3_LogP = subpc_5['AlogP'].quantile(0.75)
IQR_LogP = Q3_LogP - Q1_LogP

Q1_Molecular_Weight = subpc_5['Molecular_Weight'].quantile(0.25)
Q3_Molecular_Weight = subpc_5['Molecular_Weight'].quantile(0.75)
IQR_Molecular_Weight = Q3_Molecular_Weight - Q1_Molecular_Weight

## Remove Outlier
subpc_6 =   subpc_5[subpc_5.RO5_Violations.between(Q1_RO5-1.5*IQR_RO5, Q3_RO5+1.5*IQR_RO5, inclusive=True)]
subpc_7 =   subpc_6[subpc_6.AlogP.between(Q1_LogP-1.5*IQR_LogP, Q3_LogP+1.5*IQR_LogP, inclusive=True)]
subpc_8 =   subpc_7[subpc_7.Molecular_Weight.between(Q1_Molecular_Weight-1.5*IQR_Molecular_Weight, Q3_Molecular_Weight+1.5*IQR_Molecular_Weight, inclusive=True)]

afterprocess_6 = subpc_8
```

```python
print(afterprocess_6.shape)
afterprocess_6.head(3)
```

```python
## PROCESS 7
## Remove Salt & Select best pX if SMILES is alike

## Impliment Function
from rdkit.Chem.SaltRemover import SaltRemover
from rdkit.Chem import MolFromSmiles,MolToSmiles

def clean_smiles (ListSMILEs):
    remover = SaltRemover()
    len(remover.salts)

    SMILES_desalt = []
    defect = []
    a = 0
    
    for i in ListSMILEs:
        mol = MolFromSmiles(i) 
        if mol != None:    
            mol_desalt = remover.StripMol(mol)
            mol_SMILES = MolToSmiles(mol_desalt)
            SMILES_desalt.append(mol_SMILES)
        else:
            a+=1
            defect.append(i)
    print(a)
    for j in defect:
        print(j)
    return SMILES_desalt

subpc = afterprocess_6

subpc_1 = subpc[(subpc['Standard_Type']=='IC50')]
subpc_1['SMILES_NS'] = clean_smiles(subpc_1.Smiles)
subpc_1.sort_values(by=['SMILES_NS','pChEMBL_Value'], inplace = True)

subpc_2 = subpc_1.drop(subpc_1[(subpc_1.SMILES_NS==subpc_1.shift(-1).SMILES_NS)].index)

afterprocess_7 =  subpc_2
```

```python
print(afterprocess_7.shape)
afterprocess_7.head(3)
```

```python
# Calculate pIC50, pKi, pKd and split in to 2 group

import math
from math import log10, inf

# split upper and lower from range
def split_val(v):
    v['lower'],v['upper'] = v["Standard_Value"].split(" - ")
    return v

def pX(df0):
    normal    = df0
    
    # normal values
    normal['Standard_Value'] = pd.to_numeric(normal['Standard_Value'])
    pX          = []
    QValues_nM  = []

    for i,j,mv,mx in zip(normal.Standard_Value, 
                      normal.Standard_Units, 
                      normal.Molecular_Weight,
                      normal.pChEMBL_Value):
        if j in ['nM','nmol/l']:
            molar = float(i)*(10**-9) # Converts nM to M
            QValues_nM.append(float(i))
            pX.append(-(math.log10(molar) if molar != 0 else -inf))
            #df0['pIC50'] = (-log(molar) if molar != 0 else -inf)
        elif j == 'pM':
            molar = float(i)*(10**-12) # Converts pM to M
            QValues_nM.append(molar/(10**-9))
            pX.append(-(math.log10(molar) if molar != 0 else -inf))
        elif j == 'mM':
            molar = float(i)*(10**-3) # Converts pM to M
            QValues_nM.append(molar/(10**-9))
            pX.append(-(math.log10(molar) if molar != 0 else -inf))
        elif j in ['muM' , 'mumol/l']:
            molar = float(i)*(10**-6) # Converts micro M to M
            QValues_nM.append(molar/(10**-9))
            pX.append(-(math.log10(molar) if molar != 0 else -inf))
        elif j in ['M' , 'mol/l']:
            molar = float(i) # Converts M to M
            QValues_nM.append(molar/(10**-9))
            pX.append(-(math.log10(molar) if molar != 0 else -inf))
        elif j in ['mug/mL' , 'mug/ml']:
            molar = float(i)/float(mv)*(10**-3) # Converts micro Microgram/ml to M
            QValues_nM.append(molar/(10**-9))
            pX.append(-(math.log10(molar) if molar != 0 else -inf))
        elif j in ['g/mL']:
            molar = float(i)/float(mv)*(10**3) # Converts micro Microgram/ml to M
            QValues_nM.append(molar/(10**-9))
            pX.append(-(math.log10(molar) if molar != 0 else -inf))
        else:
            QValues_nM.append('non of this')
            pX.append(mx)
            #df0['pX'] = 'can not calculate'

    normal['pX']      = pX
    
    columns_drop = ['upper', 'lower', 'pX_lower', 'pX_upper','QValues_nM_upper','QValues_nM_lower']
    df0 = pd.concat([normal])
    
    return normal
#    return normal
afterprocess_8 = pX(afterprocess_7)

afterprocess_8["pX"] = afterprocess_8["pX"].apply(pd.to_numeric, errors='coerce')
afterprocess_8["pX"] = afterprocess_8["pX"].round(decimals=2)
```
