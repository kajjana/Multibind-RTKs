```python
import pandas as pd
import glob, os
import numpy as np
import pickle
```

```python
working = '/share/galaxy/fahsai/NSCLC'
bindingBD = working + '/datasets/bindingDB'
```

```python
df = pd.read_csv(bindingBD + "/BindingDB_All_2021.tsv", sep='\t', error_bad_lines=False)
print(df.shape)
df.head(3)
```

```python
rawdata_df = df
```

```python
rawdata_df.columns = [c.replace(': ', '_') for c in rawdata_df.columns]
rawdata_df.columns = [c.replace('(', '') for c in rawdata_df.columns]
rawdata_df.columns = [c.replace(')', '') for c in rawdata_df.columns]
rawdata_df.columns = [c.replace(' ', '_') for c in rawdata_df.columns]
rawdata_df.columns = [c.replace('/', '_') for c in rawdata_df.columns]
rawdata_df.columns = [c.replace(':', '_') for c in rawdata_df.columns]
rawdata_df.columns = [c.replace('#', '') for c in rawdata_df.columns]
```

```python
for col in rawdata_df.columns:
    print(col)
```

```python
rawdata = rawdata_df[['BindingDB_MonomerID',
                     'Ligand_SMILES',
                     'Target_Name_Assigned_by_Curator_or_DataSource',
                     'Target_Source_Organism_According_to_Curator_or_DataSource',
                     'Ki_nM',
                     'IC50_nM',
                     'Kd_nM',
                     'PDB_IDs_for_Ligand-Target_Complex',
                     'PDB_IDs_of_Target_Chain',
                      'UniProt_SwissProt_Recommended_Name_of_Target_Chain',
                      'UniProt_SwissProt_Entry_Name_of_Target_Chain',
                      'UniProt_SwissProt_Primary_ID_of_Target_Chain',
                      'UniProt_SwissProt_Secondary_IDs_of_Target_Chain',
                      'UniProt_SwissProt_Alternative_IDs_of_Target_Chain',
                      'Curation_DataSource',
                      'Article_DOI'
                     ]]
rawdata
```

target = EGFR

```python
#Process 1
#Selection of wild-type target protein

subpc = rawdata

subpc_1 = subpc[~subpc['Target_Name_Assigned_by_Curator_or_DataSource'].str.contains('mutant', na=False)]

subpc_2 = subpc_1[(subpc_1.UniProt_SwissProt_Entry_Name_of_Target_Chain == 'EGFR_HUMAN')]
subpc_3 = subpc_2[(subpc_2.UniProt_SwissProt_Primary_ID_of_Target_Chain == 'P00533')]
subpc_4 = subpc_3[(subpc_3.Target_Name_Assigned_by_Curator_or_DataSource == 'Epidermal growth factor receptor')]

afterprocess_1 = subpc_4
```

```python
print(afterprocess_1.Target_Name_Assigned_by_Curator_or_DataSource.value_counts())
print(afterprocess_1.UniProt_SwissProt_Entry_Name_of_Target_Chain.value_counts())
print(afterprocess_1.UniProt_SwissProt_Primary_ID_of_Target_Chain.value_counts())
print(afterprocess_1.UniProt_SwissProt_Secondary_IDs_of_Target_Chain.value_counts())
```

```python
print(afterprocess_1.shape)
afterprocess_1.head(3)
```

```python
#Process2
#Selection of ligands with IC50 values

subpc = afterprocess_1

subpc_1 = subpc[~subpc['IC50_nM'].isnull()]
print(subpc_1.shape)
subpc_2 = subpc_1[~subpc['IC50_nM'].str.contains('>|<|~', na=False)]
subpc_2['IC50_nM'] = pd.to_numeric(subpc_2['IC50_nM'])

subpc_3 = subpc_2.drop(['Ki_nM','Kd_nM'], axis=1)

afterprocess_2 = subpc_3
```

```python
afterprocess_2.IC50_nM.value_counts()
```

```python
print(afterprocess_2.shape)
afterprocess_2.head(3)
```

```python
#Process3
#Delete (duplicates) if the data has both ref and IC50 the same.

subpc = afterprocess_2

subpc_1 = subpc.drop_duplicates(subset = ["Ligand_SMILES","IC50_nM","Curation_DataSource","Article_DOI"], keep='first')

afterprocess_3 = subpc_1
```

```python
print(afterprocess_3.shape)
afterprocess_3.head(3)
```

```python
afterprocess_3.Ligand_SMILES.value_counts()
```

```python
## PROCESS 4
## IF there has various experiment, then data with best result is selected

subpc = afterprocess_3

subpc_1 = subpc.sort_values(by=['Ligand_SMILES','IC50_nM'], ascending=False)
subpc_2 = subpc_1.drop_duplicates(subset=['Ligand_SMILES'], keep='last')

afterprocess_4 = subpc_2
```

```python
afterprocess_4.Ligand_SMILES.value_counts()
```

```python
print(afterprocess_4.shape)
afterprocess_4.head(3)
```

```python
## PROCESS 5
## Remove Salt & Select best pX if SMILES is alike

## Impliment Function
from rdkit.Chem.SaltRemover import SaltRemover
from rdkit.Chem import MolFromSmiles,MolToSmiles

def clean_smiles (ListSMILEs):
    remover = SaltRemover()
    len(remover.salts)

    SMILES_desalt = []
    defect = []
    a = 0
    
    for i in ListSMILEs:
        mol = MolFromSmiles(i) 
        if mol != None:    
            mol_desalt = remover.StripMol(mol)
            mol_SMILES = MolToSmiles(mol_desalt)
            SMILES_desalt.append(mol_SMILES)
        else:
            a+=1
            defect.append(i)
            SMILES_desalt.append('NaN')
    print(a)
    for j in defect:
        print(j)
    return SMILES_desalt

subpc = afterprocess_4

subpc_1 = subpc[subpc.Ligand_SMILES.isnull()==False]
subpc_1['SMILES_NS'] = clean_smiles(subpc_1.Ligand_SMILES)
subpc_2 = subpc_1[~(subpc_1.SMILES_NS == 'NaN')]
subpc_3 = subpc_2.sort_values(by=['SMILES_NS','IC50_nM'], ascending=True)

subpc_4 = subpc_3.drop(subpc_3[(subpc_3.SMILES_NS==subpc_3.shift(-1).SMILES_NS)].index)

afterprocess_5 =  subpc_4
```

```python
print(afterprocess_5.shape)
afterprocess_5.head(3)
```

```python
import math
from math import log10, inf

pX = []

for ic in afterprocess_5['IC50_nM']:
    molar = float(ic)*(10**-9)
    pX.append(-(math.log10(molar) if molar != 0 else -inf))

afterprocess_5['pX'] = pX
afterprocess_5["pX"] = afterprocess_5["pX"].apply(pd.to_numeric, errors='coerce')
afterprocess_5["pX"] = afterprocess_5["pX"].round(decimals=2)
```

```python
print(afterprocess_5.shape)
afterprocess_5.head(3)
```

```python
afterClean = afterprocess_5[['BindingDB_Reactant_set_id', 'SMILES_NS', 'pX']]
afterClean_df = pd.merge(afterClean, rawdata_df, left_index=True, right_index=True)
afterClean_df = afterClean_df.drop(['BindingDB_Reactant_set_id_y'], axis=1)
afterClean_df = afterClean_df.rename(columns={'BindingDB_Reactant_set_id_x': "BindingDB_Reactant_set_id"})
afterClean_df
```

```python
afterClean_dup = afterClean_df[afterClean_df.duplicated(subset=['SMILES_NS'])]
afterClean_dup
```

```python
afterClean_df.Curation_DataSource.value_counts()
```

```python
save_dir = bindingBD + '/ic50'
afterClean_df.to_csv(save_dir+'/All_data/egfr_ic50_BindingDB_all.csv' ,index=False)
```

```python
df = afterClean_df[['BindingDB_Reactant_set_id', 'SMILES_NS', 'pX', 'UniProt_SwissProt_Entry_Name_of_Target_Chain']]
print(df.shape)
df = df.drop_duplicates(subset=['SMILES_NS'])
df = df.rename(columns={"BindingDB_Reactant_set_id": "ID", "UniProt_SwissProt_Entry_Name_of_Target_Chain":"Target_name"})
print(df.shape)
df.head()
```

```python
df.to_csv(save_dir+'/Prepared_data/egfr_ic50_BindingDB_prepared.csv' ,index=False)
```
